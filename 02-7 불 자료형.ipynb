#### 자료형의 True 와 False
- 'python' : True
- ""       : False
- [1,2,3]  : True
- []       : False
- (1,2,3)  : True
- ()       : False
- {'a':1}  : True
- {}       : False
- 1        : True
- 0        : False
- None     : False

```python
bool("")
```

```python
bool(" ")
```

```python
a = [1,2,3,4]
while a:
    print(a.pop())
```

```python
print(a)
```

```python
a = [1,2,3,4]
while a:
    print(bool(a))
    print(a)
    print(a.pop())
    print(a)
```

```python
a = [1,2,3,4]

for i in range(10):
        print(bool(a))
        # print(a)
        print(a.pop())
        print('\n')
        print(a)
        
```
