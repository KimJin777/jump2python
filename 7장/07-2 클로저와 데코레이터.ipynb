```python
class Mul:
    def __init__(self, m):
        self.m = m

    def __call__(self, n):
        return self.m*n


if __name__ =='__main__':
    mul3 = Mul(3)
    mul5 = Mul(5)
```

```python
print(mul3(10))
print(mul5(10))
```

```python
def mull(m):
    def wrapper(n):
        return m*n
    return wrapper

if __name__ =='__main__':
    mull3 = mull(3)
    mull5 = mull(5)
```

```python
print(mull3(10))
print(mull5(10))
```

```python
import time
import itertools

def elapsed(original_func):
        
    def wrapper():
        start = time.time()
        print('1함수가 실행됩니다.')
        result = original_func()
        print('2함수가 실행됩니다.')
        end = time.time()
        print('함수 수행시간: %f 초' % (end - start))
        print(result)
        return result

    return wrapper

@elapsed
def myfunc():
    print('함수가 실행됩니다.')

@elapsed
def mylucky():
    print('a')
    l = list(range(1, 45))
    print(l)
    it = itertools.combinations(l, 4)
    print(list(it))
    return it
```

```python
import time
import itertools

def elapsed(original_func):
        
    def wrapper(*args, **kwargs):
        start = time.time()
        result = original_func(*args, **kwargs)
        end = time.time()
        print('함수 수행시간: %f 초' % (end - start))
        return result

    return wrapper

@elapsed
def myfunc(msg):
    print(' %s 함수가 실행됩니다.' %msg)
```
