```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
a.setdata(4,2)
```

```python
print(a.add())
print(a.mul())
print(a.sub())
print(a.div())
```

```python
b = FourCal()
b.setdata(9,3)
```

```python
print(b.add())
print(b.mul())
print(b.sub())
print(b.div())
```

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal(8, 4) #     .
a.add()
```

```python
print(a.first, a.second)
```

```python
print(a.add())
print(a.mul())
print(a.sub())
print(a.div())
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
```

```python
c = MoreFourCal(8, 4)
c.add()
```

```python
print(c.add())
print(c.mul())
print(c.sub())
print(c.div())
```

```python
c.pow()
```

```python
c = MoreFourCal(8, 0)
c.div()
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result

    def div(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
```

```python
c = MoreFourCal(8, 0)
c.div()
```
